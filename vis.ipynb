```python
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
zcases = ['zsquares', 'zline', 'zconst']
bcases = ['blinear', 'bzero']
gcases = ['gconst', 'ggauss']

for zcase in zcases:
    for bcase in bcases:
        for gcase in gcases:
            case_root = "./ref-solutions/heat-1T/" + '-'.join([zcase, bcase, gcase])
            if not os.path.exists(case_root):
                os.makedirs(case_root)
```

```python
data1 = np.loadtxt('./ref-solutions/heat-1T/zconst-bzero-ggauss/T1_5.0e-3-1.0e-3.txt')
data2 = np.loadtxt('./ref-solutions/heat-1T/zconst-bzero-ggauss/T1_5.0e-3-1.0e-5.txt')
plt.subplot(131)
plt.imshow(data1[:,2].reshape(129,129), cmap='jet')
plt.colorbar()
plt.subplot(132)
plt.imshow(data2[:,2].reshape(129,129), cmap='jet')
plt.colorbar()
plt.subplot(133)
plt.imshow(np.abs(data1[:,2] - data2[:,2]).reshape(129,129), cmap='jet')
plt.colorbar()
plt.tight_layout()
rl2 = np.linalg.norm(data1[:,2] - data2[:,2])/np.linalg.norm(data2[:,2])
print("{:.4e}".format(rl2))
```
